# 🚀 Multi-Agent AI System for Company Valuation

## Enterprise-Grade Financial Analysis with Multi-Agent AI

This system demonstrates a sophisticated approach to company valuation by combining **three specialized AI agents** with real-time macroeconomic data.

**Key Innovation**: Instead of relying on a single valuation method, our multi-agent system creates consensus through:
- **DCF Agent** – Traditional discounted cash flow analysis
- **Risk Agent** – Company-specific risk assessments
- **Macro Agent** – Macroeconomic impact analysis

*Watch as the system analyzes major tech companies and provides confidence-scored valuations.*

---

## 🔧 System Architecture & Capabilities

| Component | Responsibility | Data Source |
|-----------|-----------------|-------------|
| **DCF Agent** | Discounted Cash Flow modeling | Financial statements (yfinance) |
| **Risk Agent** | Volatility & risk adjustments | Market data analysis |
| **Macro Agent** | Economic factor integration | FRED (inflation, GDP, unemployment) |
| **Consensus Engine** | Synthesizes all perspectives | Agent outputs + weighting |
| **Stability Testing** | Validates robustness | Scenario analysis |

**Technology Stack**: Python, LangChain, OpenAI GPT, FRED API, yfinance

````python
# ==========================================
# INITIALIZATION: Load Dependencies & APIs
# ==========================================

import os
import json
from dotenv import load_dotenv
import pandas as pd
from datetime import datetime

# Load environment variables from .env
load_dotenv()

# Import the main pipeline
from pipeline import run_pipeline

# ==========================================
# CONFIGURATION & DISPLAY UTILITIES
# ==========================================

# Verify API configurations
fred_key = os.getenv("FRED_API_KEY")
openai_key = os.getenv("OPENAI_API_KEY")

print("\n" + "="*80)
print("🔐 SYSTEM CONFIGURATION CHECK")
print("="*80)
print(f"✅ FRED API Key:   {('Configured' if fred_key else '❌ NOT CONFIGURED')}")
print(f"✅ OpenAI API Key: {('Configured' if openai_key else '❌ NOT CONFIGURED')}")
print(f"📍 Working Dir:    {os.getcwd()}")
print(f"⏰ Timestamp:      {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
print("="*80 + "\n")

if not (fred_key and openai_key):
    print("⚠️  WARNING: Missing API keys. Please ensure .env file contains:")
    print("   - FRED_API_KEY (from https://fredaccount.stlouisfed.org/apikey)")
    print("   - OPENAI_API_KEY (from https://platform.openai.com/api-keys)")
    print()

# ==========================================
# DISPLAY UTILITIES
# ==========================================

def display_valuation(ticker, result):
    """Beautiful display of valuation results."""
    print(f"\n{'█'*80}")
    print(f"{'█'} VALUATION RESULTS: {ticker}{' '*(80-len(f'█ VALUATION RESULTS: {ticker}'))}")
    print(f"{'█'*80}")
    
    if isinstance(result, dict):
        for key, value in result.items():
            if isinstance(value, (int, float)):
                print(f"  {key:.<40} ${value:>15,.2f}")
            else:
                print(f"  {key:.<40} {str(value)[:50]:>15}")
    else:
        print(json.dumps(result, indent=2))
    
    print(f"{'█'*80}\n")

print("✅ System ready. Pipeline loaded successfully.")
print("\n💡 Tip: Each company valuation takes 30-60 seconds as AI agents analyze data.\n")
````

## 📊 LIVE DEMO: Valuing Real Companies

### Example 1️⃣ – Apple Inc. (AAPL)

````python
print("\n⏳ Analyzing Amazon (AMZN)... Running 3-agent valuation pipeline...\n")
result_amzn = run_pipeline("AMZN")
display_valuation("AMZN", result_amzn)
````

---

## 📈 Comparative Analysis: Side-by-Side Valuation Comparison

````python
print("\n📊 GENERATING COMPARATIVE ANALYSIS TABLE...\n")

# Create structured comparison - Clean and simple
comparison_data = []

for ticker, result in [("AAPL", result_aapl), ("V", result_v), ("AMZN", result_amzn)]:
    if isinstance(result, dict):
        consensus_val = result.get('final_consensus')
        consensus_str = f"${consensus_val/1e9:.2f}B" if isinstance(consensus_val, (int, float)) else "N/A"
        
        comp_row = {
            "Company": ticker,
            "Final Valuation": consensus_str,
        }
    else:
        comp_row = {
            "Company": ticker,
            "Final Valuation": "N/A",
        }
    comparison_data.append(comp_row)

df = pd.DataFrame(comparison_data)

print("\n" + "="*70)
print("COMPARATIVE VALUATION SUMMARY – CONSENSUS VALUES")
print("="*70)
print(df.to_string(index=False))
print("="*70)
print("\n✅ All 3 companies successfully analyzed by multi-agent system")
print()
````
